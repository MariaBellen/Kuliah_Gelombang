Seperti kalkulator, sudut yang dimasukkan dalam operasi sinus adalah sudut dalam radian, jadi sudut derajat harus dikali dengan $\frac{\pi}{180}$

untuk menggambar juga mudah, kita hanya perlu perintah **plot**, namun memerlukan modul matplotlib dulu, seperti contoh berikut:
Gambarkan grafik fungsi sinus dengan frekuensi 1 Hz selama 10 detik, dengan amplitudo 2 satuan. Maka persamaannya adalah
<center>$f = A \times \sin 2 \pi f t = 2 \times \sin 2 \pi t, t = 1..10$</center>
Untuk membuat vektor waktu, $t$, diperlukan modul numpy (numerical Python), yang harus diimport terlebih dahulu

```python
import math
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0,10,0.05)
f = 2 * np.sin(2*math.pi*1*t)
plt.plot(t,f)
plt.xlabel('waktu[dt]')
plt.ylabel('Amplitudo')
plt.show()
```

Seperti definisi frekuensi, maka dalam 10 detik, sinyal 1 Hz akan terdiri atas sepuluh gelombang. Perhatikan apa yang terjadi saat frekuensinya jadi 0,5 Hz.

```python
import math
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0,10,0.05)
f = 2 * np.sin(2*math.pi*0.5*t)
plt.plot(t,f)
plt.xlabel('waktu[dt]')
plt.ylabel('Amplitudo')
plt.show()
```
